```python
data = x
size_x, size_y = data.shape
numel = size_x * size_y
```

```python
X = data.reshape(-1)
X = torch.tensor(X.copy())
```

```python
%%time
X_grid1, X_grid2 = torch.meshgrid(X, X)
M = torch.exp(-(X_grid1 - X_grid2) ** 2)
```

```python
%%time
M = torch.exp(-(X.view(-1, 1) - X) ** 2)
```

```python
%%time
d = M.sum(dim=0)
d_half = d**(-0.5)
N = torch.eye(numel) - d_half.view(-1, 1) * M * d_half
```

```python
%%time
with torch.no_grad():
    evals, evecs = torch.lobpcg(N, largest=False, k=4)
```

```python
kmeans = KMeans(n_clusters=4, n_init=1)
```

```python
%%time
pred = kmeans.fit(evecs.numpy())
```

```python
def score(pred, targ, basic=True):
    assert pred.shape == targ.shape
    shape = pred.shape
    pred = pred.copy().reshape(-1)
    targ = targ.copy().reshape(-1)
    pred_labels = np.unique(pred)
    pred_masks = [pred == x for x in pred_labels]
    pred_perms = itertools.permutations(pred_labels)
    best_perm = None
    best_score = 0
    for pred_perm in pred_perms:
        for i, label in enumerate(pred_perm):
            pred[pred_masks[i]] = label
        if basic:
            score = np.mean(pred[targ <= 3] == targ[targ <= 3])
        else:
            score = np.mean(pred == targ)
        if score > best_score:
            best_score = score
            best_perm = pred.copy()
    return best_score, best_perm.reshape(shape)
```

```python
val, perm = score(pred.labels_.reshape(size_x, size_y), y, basic=True)
```
